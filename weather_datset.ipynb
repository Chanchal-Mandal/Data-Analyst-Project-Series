```python
# Dropping the rows with NaN values

import pandas as pd
df = pd.read_csv(r"C:\Users\chanc\Downloads\weather.csv")
df.dropna(inplace =True)
df.isnull().sum()
```

```python
# Converting rain today and tommorow categorical value to numeric value

df["RainToday"] = df["RainToday"].map({
    "No" : 0,
    "Yes" : 1
}.get)


df.head()
```

```python
df["RainTomorrow"] = df["RainTomorrow"].replace({
    "No" : 0,
    "Yes" : 1
})
df.head()
```

```python
df.WindGustDir.unique()
```

```python
df.WindGustDir.replace({
    'NW' : 1,
    'ENE' : 2,
    'SSE' : 3,
    'SE' : 4,
    'E' : 5,
    'S' : 6,
    'N' : 7,
    'WNW': 8,
    'ESE' : 9,
    'NE' : 10,
    'NNE' : 11,
    'NNW' : 12,
    'SW' : 13,
    'W' : 14,
    'WSW' : 15,
    'SSW' : 16,
}, inplace = True)

df.WindDir3pm.replace({
    'NW' : 1,
    'ENE' : 2,
    'SSE' : 3,
    'SE' : 4,
    'E' : 5,
    'S' : 6,
    'N' : 7,
    'WNW': 8,
    'ESE' : 9,
    'NE' : 10,
    'NNE' : 11,
    'NNW' : 12,
    'SW' : 13,
    'W' : 14,
    'WSW' : 15,
    'SSW' : 16,
}, inplace = True)

df.WindDir9am.replace({
    'NW' : 1,
    'ENE' : 2,
    'SSE' : 3,
    'SE' : 4,
    'E' : 5,
    'S' : 6,
    'N' : 7,
    'WNW': 8,
    'ESE' : 9,
    'NE' : 10,
    'NNE' : 11,
    'NNW' : 12,
    'SW' : 13,
    'W' : 14,
    'WSW' : 15,
    'SSW' : 16,
}, inplace = True)

df.head()
```

```python
df['Pressure9am'] = (df['Pressure9am'] - df['Pressure9am'].min()) / (df['Pressure9am'].max() - df['Pressure9am'].min())
df.head()
```

```python
df['Pressure3pm'] = (df['Pressure3pm'] - df['Pressure3pm'].min()) / (df['Pressure3pm'].max() - df['Pressure3pm'].min())
df.head()
```

```python
df.to_csv("C:/Users/chanc/Downloads/preprocessed.csv",index = False)
```
